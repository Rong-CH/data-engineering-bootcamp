# Question 2.
What are the top 10 monitored products by State?

```python
for estados in df['estado'].unique():
    estado = estados                   #grabs the state
    uniq1 = df.loc[df['estado'] == estado]['marca'] #obtain marca from each state
    keys_list = Counter(uniq1).keys() # equals to list(set(words))
    values_list = Counter(uniq1).values() # counts the elements' frequency
    zip_iterator = zip(keys_list, values_list) #concatenate both lists
    a_dictionary = dict(zip_iterator) # create a dictionary from keys and values
    sor_dict = sorted(a_dictionary, key=a_dictionary.get, reverse=True) #sort values from the highest value
    print("------------------------")
    print(estado)
    print("------------------------")
    for s in sor_dict[1:11]: # present the ten highest keys (ignoring S/M)
        print(s)
```

# Question 3.
Which is the commercial chain with the highest number of monitored products?

```python
comercial = []                    #declaring empty lists
n_productos = []            
for cadena in tqdm(df['cadenaComercial'].unique()):         #populating empty lists 
    cad1 = df.loc[df['cadenaComercial'] == cadena]['producto']
    comercial.append(cadena)
    n_productos.append(int(len(cad1)))
    
```

```python
zip_iterator = zip(comercial, n_productos)                 #If one tuple contains more items these items are ignored
a_dictionary = dict(zip_iterator)
sor_dict = sorted(a_dictionary, key=a_dictionary.get, reverse=True) #sort comercial
max_key = max(a_dictionary, key=a_dictionary.get) #max value
print(max_key)
```

```python
for com in sor_dict[:10]:
    print(com)
```
